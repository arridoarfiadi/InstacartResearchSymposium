```python
from sklearn.externals import joblib
import pandas as pd
import numpy as np
import pickle
pd.set_option('display.max_rows', 500)
```

```python
predictions = joblib.load("FinalReorderProbability.dms")
```

```python
with open('orig_products.pickle', "rb") as f:
    products = pickle.load(f)
```

```python
predictionsWithProductName = predictions.merge(products, on = ["ProductID"], how = "left")
```

```python
def getUserProductProbability(userID):
    return predictionsWithProductName[predictionsWithProductName.UserID == userID].sort_values(by = "ReorderProbability", ascending = False)[["UserID","ProductName","ReorderProbability"]]
def getProductReorderProbability(productID):
    return predictionsWithProductName[predictionsWithProductName.ProductID == productID].sort_values(by = "ReorderProbability", ascending = False)[["UserID","ProductName","ReorderProbability"]]
```
